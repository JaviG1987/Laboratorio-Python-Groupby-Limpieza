```python
import pandas as pd
pd.set_option ("display.max_columns", None)
```

```python
df_mergeado_netflix = pd.read_csv("data_mergeado.csv", index_col=0)
df_mergeado_netflix.head(1)
```

#### Ejercicio 1: ¿Cuántos títulos hay por cada tipo de contenido ('Movie' o 'TV Show')?

```python
numero_titulos =df_mergeado_netflix.groupby("type")["Title"].count().reset_index()
numero_titulos
```

#### Ejercicio 2: ¿Cuál es el promedio de puntuación de las películas agrupadas por clasificación de audiencia ('rating')? Interpreta los resultados, ¿hay algo que te llame la atención? Explica por qué. 

```python
promedio_puntuacion = df_mergeado_netflix.groupby("rating")["IMDB Score"].mean().reset_index().round(2)
promedio_puntuacion
```

#### Ejercicio 3: ¿Cuántos títulos hay por cada país? Ordena los resultados de mayor a menor

```python
titulos_pais = df_mergeado_netflix.groupby("country")["Title"].count().reset_index()
titulos_pais_ordenado = titulos_pais.sort_values("Title", ascending=False)
titulos_pais_ordenado
```

#### Ejercicio 4: ¿Cuál es la calificación promedio en IMDb de los títulos agrupados por tipo ('Movie' o 'TV Show')?

```python
tipo_titulos = df_mergeado_netflix.groupby("type")["IMDB Score"].mean().reset_index().round(2)
tipo_titulos
```

#### Ejercicio 5: ¿Cuántos títulos se añadieron a Netflix cada año? Para resolver este ejercicio tendrás uqe convertir la columna `date_added` a formato fecha usando el método `pd.to_datetime`. [Aquí](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html) tienes la documentación, leeala con detenimiento antes de usar el método. 

```python
df_mergeado_netflix["date_added"] = pd.to_datetime(df_mergeado_netflix["date_added"],errors="coerce")
```

```python
df_mergeado_netflix.info()
```

```python
df_mergeado_netflix ["date_added_año"] = df_mergeado_netflix ["date_added"].dt.year
titulos_año = df_mergeado_netflix.groupby("date_added_año")["Title"].count().reset_index()
titulos_año
```

#### Ejercicio 6: ¿Cuál es el país con más títulos lanzados en el año 2020? Para resolver este ejercicio después del gropuby os puede resultar interesante el método [`idxmax`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.idxmax.html)

```python
filtro_2020 = df_mergeado_netflix ["date_added_año"] == 2020.0
titulos_pais_2020 = df_mergeado_netflix[filtro_2020].groupby("country")["Title"].count().reset_index()
titulos_pais_2020_ordenado = titulos_pais_2020.sort_values ("Title", ascending = False)
titulos_pais_2020_ordenado
```

```python
maximo_titulos_2020 = titulos_pais_2020_ordenado["Title"].idxmax()
fila_max = titulos_pais_2020_ordenado.loc[maximo_titulos_2020]
df_maximo_titulos_2020 = pd.DataFrame(fila_max)
df_maximo_titulos_2020
```

#### Ejercicio 7:¿Cuál es el promedio de puntuación en IMDb de los títulos agrupados por país?

```python
promedio_pais = df_mergeado_netflix.dropna(subset=["IMDB Score"]).groupby ("country")["IMDB Score"].mean().reset_index().round(2)
promedio_pais
```

#### Ejercicio 8: ¿Cuántos títulos fueron lanzados en cada año agrupados por tipo ('Movie' o 'TV Show')?

```python
titulos_lanzados = df_mergeado_netflix.groupby(["date_added_año","type"])["Title"].count().reset_index()
df_cambio_filas = titulos_lanzados.unstack()
df_cambio_filas
```

#### Ejercicio 9: ¿Cuál es el país con la mayor cantidad de títulos de clasificación 'TV-MA'?

```python
filtro_TVMA = df_mergeado_netflix["rating"] == "TV-MA"
pais_TVMA = df_mergeado_netflix[filtro_TVMA].groupby ("country")["Title"].count()
pais_TVMA_maximo = pais_TVMA.idxmax()
print(f"El pais con mayor cantidad de titulos TV-MA es {pais_TVMA_maximo}")
```

#### Ejercicio 10. ¿Cuál es el promedio de duración de las películas agrupadas por país?

```python
duracion_pelis_pais = df_mergeado_netflix.groupby("country")["Runtime"].mean().reset_index().round(2)
duracion_pelis_pais
```

#### Ejercicio 11: ¿Cuál es el número total de títulos por país y clasificación de audiencia ('rating')?

```python
titulos_pais_clasificacion = df_mergeado_netflix.groupby (["country","rating"])["Title"].count().reset_index()
titulos_pais_clasificacion
```

```python
tabla_titulos_pais_clasificacion = titulos_pais_clasificacion.pivot(index="country", columns="rating", values="Title").fillna(0)
tabla_titulos_pais_clasificacion
```

#### Ejercicio 12: ¿Cuántos títulos fueron lanzados antes del año 2000 agrupados por país?

#### Ejercicio 13: ¿Cuál es la puntuación promedio en IMDb de los títulos lanzados en 2021 agrupados por país?

```python
filtro_2021 = df_mergeado_netflix["date_added_año"] == 2021.0
df_IMBD_sinnulos = df_mergeado_netflix.dropna(subset=["IMDB Score"])
puntuacion_titulos2021 = df_IMBD_sinnulos[filtro_2021].groupby("country")["IMDB Score"].mean().reset_index().round(2)
puntuacion_titulos2021
```

#### Ejercicio 14: ¿Cuántos títulos fueron dirigidos por cada director? Ordena los resultados de mayor a menor

#### Ejercicio 15: ¿Cuál es el lenguaje más frecuente en los títulos lanzados en los últimos 5 años agrupado por país?

```python
from datetime import datetime
año_actual = datetime.now().year
filtro_ultimos5 = df_mergeado_netflix["date_added_año"] >= año_actual - 5
lenguaje_ultimos5 = df_mergeado_netflix[filtro_ultimos5].groupby(["country","date_added_año"])["Language"].count().reset_index()
lenguaje_ultimos5_ordenado = lenguaje_ultimos5.sort_values("Language", ascending=False)
lenguaje_ultimos5_ordenado
```
